```python
#TODO 1:
from PIL import Image
import matplotlib.pyplot as plt

# Load the image file into a PIL Image object
img_pil = Image.open("image.jpg")

# Display basic information about the image
print("Image object type:", type(img_pil))
print("Image size (W, H):", img_pil.size)
print("Image color mode:", img_pil.mode)

# Render the image using matplotlib
plt.imshow(img_pil)
plt.axis("off")
```

```python
#TODO 2:
import numpy as np

# Transform the PIL image into a NumPy array
img_np = np.asarray(img_pil)

# Show array structure and data type
print("Array dimensions:", img_np.shape)
print("Array data type:", img_np.dtype)
```

```python
#TODO 3:
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Read image file directly as a NumPy array
img_mpl = mpimg.imread("image.jpg")

# Output the size of the image array
print("Image array shape:", img_mpl.shape)

# Display the image
plt.imshow(img_mpl)
plt.axis("off")
```

```python
#Why does Matplotlib treat images as NumPy arrays?

#Matplotlib is built on NumPy and is designed for numerical visualization. Treating images as NumPy arrays allows direct mathematical operations, slicing, and integration with scientific computing workflows.
```

```python
#TODO 4:#Why does Matplotlib treat images as NumPy arrays?

#Matplotlib is built on NumPy and is designed for numerical visualization. Treating images as NumPy arrays allows direct mathematical operations, slicing, and integration with scientific computing workflows.
from skimage import io, img_as_float
import matplotlib.pyplot as plt

# Load the image using scikit-image
img_sk = io.imread("image.jpg")

# Convert pixel values to floating-point range (0 to 1)
img_float = img_as_float(img_sk)

# Print minimum and maximum pixel values
print("Lowest pixel value:", img_float.min())
print("Highest pixel value:", img_float.max())

# Show the image
plt.imshow(img_float)
plt.axis("off")
```

```python
#Why is astype(float) discouraged compared to img_as_float?

#astype(float) doesn't rescale pixel values and can lead to incorrect intensity ranges.
#img_as_float converts and normalizes image data according to its original type, keeping visual correctness.
```

```python
#TODO 5:
import cv2
import matplotlib.pyplot as plt

# Read the image as grayscale
img_gray = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)

# Read the image in color (OpenCV uses BGR format)
img_bgr = cv2.imread("image.jpg")

# Convert BGR color format to RGB for correct display
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Display the corrected RGB image
plt.imshow(img_rgb)
plt.axis("off")
```

```python
#TODO 6:
# Perform edge detection on the grayscale image
edges = cv2.Canny(
    img_gray,       # Input image
    100,            # Lower threshold
    200             # Upper threshold
)

# Show the edge-detected result
plt.imshow(edges, cmap="gray")
plt.axis("off")
```

```python
#BONUS TASK B:
import glob
import cv2
import matplotlib.pyplot as plt

# Collect all JPG files from the images directory
files = glob.glob("images/*.jpg")

# Process each image file one by one
for file in files:
    # Load image using OpenCV
    img = cv2.imread(file)

    # Convert color channels from BGR to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display image without axes
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```
