```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('/Users/230271/Desktop/bmw_m8.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
subtracted_image = cv2.subtract(image_RGB, 100)
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(subtracted_image)
plt.title('subtracted Image (-100)')
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/Users/230271/Desktop/bmw_m8.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

result = image_RGB.copy()


result[:, :, 0] = cv2.subtract(result[:, :, 0], 100)

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title("original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(result)
plt.title("red ")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/Users/230271/Desktop/bmw_m8.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

multiplied_image = cv2.multiply(image_RGB, 2)

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title("original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(multiplied_image)
plt.title("multiplied ")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/Users/230271/Desktop/bmw_m8.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

divided_image = cv2.divide(image_RGB, 2)

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title("original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(divided_image)
plt.title("divided ")
plt.axis("off")

plt.show()
```
